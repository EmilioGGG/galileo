```python
# Importando las libreríass a utilizar
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from sklearn.preprocessing import PolynomialFeatures
import warnings
warnings.filterwarnings("ignore")
```

```python
data_total = np.load('C:/Users/Gary/Google Drive/Universidad/MAESTRIA/Data Science/Año 1/3. Trimestre III/2. Statistical Learning II/Laboratorios/Lab 3/proyecto_training_data.npy')
data_final = np.nan_to_num(data_total,copy = True, nan = 0.0)
```

```python
print("shape:", data_final.shape)
```

```python
data_probar = data_final[:,1:5]
```

```python
data_probar.shape
```

```python
fig, ae = plt.subplots(2,2, figsize=(18,10))
sns.distplot(data_probar[:,0],ax=ae[0,0])
sns.distplot(data_probar[:,1],ax=ae[0,1])
sns.distplot(data_probar[:,2],ax=ae[1,0])
sns.distplot(data_probar[:,3],ax=ae[1,1])
```

# Polinomio Grado 2

```python
poly = PolynomialFeatures(degree=2)
data_final2 = poly.fit_transform(data_probar)
```

```python
data_final2.shape
```

```python
fig, ae = plt.subplots(5,3, figsize=(18,10))
sns.distplot(data_final2[:,0],ax=ae[0,0])
sns.distplot(data_final2[:,1],ax=ae[0,1])
sns.distplot(data_final2[:,2],ax=ae[0,2])
sns.distplot(data_final2[:,3],ax=ae[1,0])
sns.distplot(data_final2[:,4],ax=ae[1,1])
sns.distplot(data_final2[:,5],ax=ae[1,2])
sns.distplot(data_final2[:,6],ax=ae[2,0])
sns.distplot(data_final2[:,7],ax=ae[2,1])
sns.distplot(data_final2[:,8],ax=ae[2,2])
sns.distplot(data_final2[:,9],ax=ae[3,0])
sns.distplot(data_final2[:,9],ax=ae[3,1])
sns.distplot(data_final2[:,9],ax=ae[3,2])
sns.distplot(data_final2[:,9],ax=ae[4,0])
sns.distplot(data_final2[:,9],ax=ae[4,1])
sns.distplot(data_final2[:,9],ax=ae[4,2])
```

```python
#Creando un array de 1D para poder crear un plot comparativo
a_poly = data_final2.flatten()
xarr = np.arange(1,a_poly.size+1,1)
plt.scatter(xarr,a_poly)
plt.title("Degree 2 Polynomial")
plt.show()
```
