```python
import numpy as np
```

```python
def sigmoid(X):
  return 1/(1+np.exp(-X))

sigmoid(np.array([2,3,4]))
```

```python
#Matriz de parametros para un modelo de 3 neuronas
Wc1 = np.array([
    [0.9,0.3,0.2],
    [0.8,0.5,0.4],
    [0.1,0.6,0.7]
])
```

```python
Wc2 = np.array([0.3,0.5,0.9])
```

```python
# 4 observaciones, 3 variables
X = np.array([
      [1,0,1],
      [1,2,3],
      [4,5,6],
      [6,7,8],
])
```

### Forward-propagation(inferencia, prediccion, o estimacion)

```python
H = np.matmul(X,Wc1) #1. calcular salida en la primer direccion #Producto punto entrefilas y columnass
```

```python
class RedNeuronal1:
  def __init__(self):
    # 1 fila por neurona , 1 columna por variable de entrada
    self.Wc1 = np.array([
        [0.8,0.2],
        [0.4,0.9],
        [0.3,0.5]    
    ])

    # 1 elemento para cada neurona intermedia
    self. Wc2  = np.array([0.3,0.5,0.9])

  def forward_propagation1(self,X):
    print(X.shape)
    print(self.Wc1.T.shape)
    A = sigmoid(np.matmul(X,self.Wc1.T)) # 1 fila por observacion, 1 columna por representacion intermedia [4,3]
    print(A)
    print(A.shape)
    y_aprox = sigmoid(np.matmul(A,self.Wc2)) # 1 fila por observacion,  1 unica salida [4,1]
    print(y_aprox.shape)
    return y_aprox
```
