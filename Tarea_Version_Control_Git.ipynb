# Missing Semester - Version Control GIT




### Sistemas de control de versiones:
Git es uno de los sistemas más utilizados para el control de versiones, este es utilizado para mantener un orden y un seguimiento de los cambios a nuestros trabajos.
Estos sistemas ayudan  a trackear el historial de los cambios que se realizan en un proyecto, es decir, caputura los cambios por folder y todo lo que está dentro de el para saber en que momento se realizó algún cambio en especifico.

Cada proyecto contiene metadata, en la cual encontramos informción de __cuando y quien realizó algún cambio en los documentos o proyectos.__

Es importante recalcar que con GIT podemos ver versiones anteriores de proyectos en los que trabajamos, con lo cual podemos podemos regresar a una versión del proyecto en caso fuera necesario.
GIT nos permite trabajar o colaborar con otras personas dentro del equipo, trabajar en el mismo proyecto al mismo tiempo sin tener que esperar a que las otras personas terminen de trabajar en el, por ejemplo se puede enviar parches del proyecto para que otra persona los pueda utilizar o resolver problemas de equipo en conjunto.

## Git Data Model:
El modelo de datos de Git es una coleccion de directorios en donde se guardan una serie de "fotografías" del proyecto, es decir, es una linea de tiempo del proyecto que guarda la información de los cambios importantes de este.

Un directorio es llamado "Arbol" o "Tree", este contiene folders y dentro de cada uno de estos se encuentran otros folders y/o archivos del proyecto.
Una forma de entender como es el modelo de datos es el siguiente:

Root  ----->Tree"Forlder" __Directorio Raíz__
...Folder1 __Puede contener otros folders o archivos__
.....arc1.txt
 ....arc2.txt
..Folder2 __Tambien puede contener folders o archivos__
-----arch1.txt
-----arch2.txt
..arch4.txt __La raíz del proyecto tambien puede almacenar archivos

GIT va creando una copia de este folder a travez del timepo, con el objetivo de entender o saber que cambipo precedió a X cambio, quien lo realizó y el porque lo realizó __(Cada vez que se hace commit es necesario escribir comentarios del cambio)__

Es importante tambien saber que del proyecto en que se está trabajando se pueden crear "bifurcaciones" es decir, copias del proyecto que apuntan a la versión anterior, esto con el objetivo de trabajar en distintos cambios del proyecto por diferentes personas, adicionalmente estos cambios se pueden volver a juntar y hacer una sola versión del proyecto nuevamente.

### Como se guarda la informacion en GIT:
La información en GIT son un set de objetos, un commit o "guardar" hace referencia a padres por sus ID's.
Cuando un objeto hace referencia a otro objeto lo hace por medio del hash.
Por ejemplo, un archivo no hace referencia por nombre del otro archivo, sino por su hash.
    

### Objetos y referencias:

Por los nombres tan largos en que se guarda la información GIT, este nos da una solución que nos permite hacer más facil el entender como se relacionan los objetos, esto lo hace por medio de referencias:
Una referencia es un mapa de string a string, leíble para el hombre para saber referenciar a algun snapshot de la historia.

## Comandos de GIT:
Es importante tomar en cuenta que el sistema GIT utiliza un command line del sistema operativo Linux, lo cual pesonalmente facilita el uso y la interacción de GIT.
Podemos utilizar comandos como 
mkdir, ls, pwd, rmdir, diff, grep, cat, etc.

### git config --global user.name
Este comando configura el nombre de usuario que está usando el enviroment.
<br>Ejemplo:
<br>
__git confir --global user.name "EmilioGGG"__

### git config --global user.email
Este comando configura el correo de usuario que está usando el enviroment.
<br>Ejemplo 
<br>__git config --global user.email "emilio.gonzalez@galileo.edu__

#### git init  
Este comando inicializa un repositorio git en la ubicación donde estamos actualmente, este nos crea archivos oculto de git en donde se guardaran objetos y referencias.
para poder ver todos los archivos que nos creó podemos utilizar el comando 
ls .git  Muestra todos los archivos extensión git.

#### git help 
Este comandonos muestra la ayuda de git, en el cual podemos guiarnos y encontrar ayuda para utilizar la herramienta

#### git status 
Este nos muestra el status del archivo o proyecto en el cual estamos trabajando, cuando no se ha realizado un commit __nos muestra en rojo el archivo que no está guardado.__

### git add archivo.ext 
Este agrega el archivo a espera para hacer commit, es decir, este archivo se agrega a la lista en la cual al hacer commit del proyecto se guardará. Importante: si no hago este comando al hacer commit no me lo agregará y no se guardarán los cambios.
<br>Ejemplo:
<br>__git add prueba.txt__


### git commit
Al realizar el commit de lo que estamos trabajanro se abre un archivo el cual __obliatoriamente tenemos que escribir algo relacionado al cambio que estamos realizando__ con el objetivo de saber que se realizó.
Este archivo lo podemos editar con VIM, por lo que utiliza los comandos de VIM, como por ejemplo
#### A para modificar el archivo
#### :wq para guardar y cerrar el archivo
#### :q para cerrar el archivo sin guardar los cambios



### git log  
Este comando nos muestra la historia del commit, quien lo realizó, y la fecha en que fue realizado.
Este comando le podemos agregar comandos adicionales para que sea visualmente máss agradable y entendible, como por ejemplo:
git log --all --graph --decorate

### git checkout
Este comando nos mueve a donde estamos viendo la linea de cambios, se tiene que mover por el nombre del commit o la direccion del mismo.es decir cambia el directorio donde estamos trabajando y mueve el puntero a donde estoy trabajando hacia el que le indico en el comando.


### git diff 
Este nos muestra que ha cambiado desde el ultimo commit, en resumen este muestra la diferencia entre los documentos, que se guardo de último versus que cambios se realizaron despues del commit.
 
### git chechout
Este deshace los cambios que hice, es decir, no guarda la informacion que modificamos y deja el archivo a como estaba en el último commit.
 
   


 ### git branch 
Como se mencionó anteriormente, el proyecto se puede dividir en varias lineas para trabajarlo, pues este comando crea una copia del proyecto apuntando a su predecesor
Por ejemplo el comando __git branch v2__
se ve cual es el master y cuales están debajo de master, 2 branches, master y cat.
 
### git merge  
Este comando une las dos lineas que salen del nodo principal __(o que dividimos con el comando anterior git branch)__

 

### git remote
Enlista todos los remotos que se tienen del proyecto, es decir, otras personas spueden estar trabajando en el proyecto, este comando nos dice quien está trabajando en el en distintos ordenadores.

### git push
Este envia los cambios en mi computador al remoto, para que alguien más pueda seguir trabajando en el proyecto.
 
### git clone 
Este crea una copia localmente del servidor remoto para trabajar en el, es decir, se trae el proyecto a a mi computador par poder modificarlo sin afectar a los demás colaboradores del proyecto.
 

git fetch -  no cambia nuestra historia local, pero si en el servitod

git pull o git merge

git push . enviando copias del local al remote

git pull hace lo mismo que fetch + push
